```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Spark SQL Example").getOrCreate()
```

```python
df1 = spark.createDataFrame([(1, "Alice", 25), (2, "Bob", 30), (3, "Charlie", 35), (4, "David", 40)], ["id", "name", "age"])
df2 = spark.createDataFrame([(1, "New York"), (2, "Los Angeles"), (3, "Chicago"), (5, "Houston")], ["id", "city"])
```

```python
df1.show()
df2.show()
```

```python
# Inner Join
inner_join_df = df1.join(df2, on="id", how="inner")
print("Inner Join Result:")
inner_join_df.show()
```

```python
# Left Outer Join
left_outer_join_df = df1.join(df2, on="id", how="left")
print("Left Outer Join Result:")
left_outer_join_df.show()
```

```python
# Right Outer Join
right_outer_join_df = df1.join(df2, on="id", how="right")
print("Right Outer Join Result:")
right_outer_join_df.show()
```

```python
# Full Outer Join
full_outer_join_df = df1.join(df2, on="id", how="outer")
print("Full Outer Join Result:")
full_outer_join_df.show()
```

```python
# Left Semi Join
left_semi_join_df = df1.join(df2, on="id", how="left_semi")
print("Left Semi Join Result:")
left_semi_join_df.show()
```

```python
# Left Anti Join
left_anti_join_df = df1.join(df2, on="id", how="left_anti")
print("Left Anti Join Result:")
left_anti_join_df.show()
```

```python
# Cross Join
cross_join_df = df1.crossJoin(df2)
print("Cross Join Result:")
cross_join_df.show()
```
