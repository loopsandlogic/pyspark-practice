```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
spark = SparkSession.builder.appName("Spark SQL Example").getOrCreate()
```

```python
# Create dataframe from row
row_data = [Row(id=1, name="Alice", age=30, salary=50000.0), 
    Row(id=2, name="Bob", age=25, salary=45000.0),
    Row(id=3, name="Charlie", age=35, salary=60000.0),
    Row(id=4, name="David", age=28, salary=55000.0),
    Row(id=5, name="Eve", age=32, salary=70000.0)]

row_df = spark.createDataFrame(row_data)
row_df.show()
```

```python
tupple_data = [(1, "Alice", 30, 50000.0),
    (2, "Bob", 25, 45000.0),
    (3, "Charlie", 35, 60000.0),
    (4, "David", 28, 55000.0),
    (5, "Eve", 32, 70000.0)]

tupple_df = spark.createDataFrame(tupple_data, ["id", "name", "age", "salary"])
tupple_df.show()
```
