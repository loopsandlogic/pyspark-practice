```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("CreateParquetFile").getOrCreate()
```

```python
# Read data from a CSV file and create a parquet file
df = spark.read.csv("data/input/orders.csv", header=True, inferSchema=True, sep="|")
df.write.mode("overwrite").parquet("data/output/orders_parquet")
```

```python
# Read the generated parquet file and show its content
parquet_df = spark.read.parquet("data/output/orders_parquet")
parquet_df.show()
```

```python
spark.stop()
```
