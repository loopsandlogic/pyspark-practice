```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Spark SQL Example").getOrCreate()
```

```python
data = [("A", "East", 100), ("A", "West", 200), ("B", "East", 300), ("B", "West", 400)]

df = spark.createDataFrame(data, ["product", "region", "sales"])

df.show()
```

```python
pivot_df = df.groupBy("product").pivot("region").sum("sales")
pivot_df.show()
```

```python
df.groupBy("product").sum("sales").withColumnRenamed("sum(sales)", "total_sales").show()
```
