```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder.appName("Spark SQL Example").getOrCreate()
```

```python
df_orders = spark.read.csv("data/input/orders.csv", header=True, inferSchema=True, sep="|")
df_orders.show(5)
```

```python
df_orders.select((col("invoiceno") + 2).alias("newinvoiceno"), col("country")).show(5)
```

```python
df_orders.selectExpr("invoiceno + 2", "country as country_name").show(5)
```
